```python
import pandas as pd
df1 = pd.read_csv("word_search.csv",encoding="gbk")
```

```python
df1 = df1[["Chinese","Dongxiang"]].drop_duplicates()
```

```python
import re
import pandas as pd
from rapidfuzz import process, fuzz



cn_list = df1["Chinese"]
dx_list = df1["Dongxiang"]

def search_by_chinese(q, top_k=3, min_score=95, scorer=fuzz.WRatio):
    matches = process.extract(q, cn_list, scorer=scorer, limit=top_k)
    out = []
    for _, score, idx in matches:
        if score < min_score:
            continue
        out.append({
            "Chinese": df1.loc[idx, "Chinese"],
            "Dongxiang": df1.loc[idx, "Dongxiang"],
            "score": score
        })
    return pd.DataFrame(out)

def search_by_dongxiang(q, top_k=3, min_score=95, scorer=fuzz.WRatio):
    matches = process.extract(q, dx_list, scorer=scorer, limit=top_k)
    out = []
    for _, score, idx in matches:
        if score < min_score:
            continue
        out.append({
            "Dongxiang": df1.loc[idx, "Dongxiang"],
            "Chinese": df1.loc[idx, "Chinese"],
            "score": score
        })
    return pd.DataFrame(out)


res_cn2dx = search_by_chinese("买")
res_dx2cn = search_by_dongxiang("a")

print(res_cn2dx)
print(res_dx2cn)
```

- **top_k**: Controls the maximum number of candidate matches returned by the search.
- **min_score**: Defines the minimum similarity threshold required for a match to be accepted, effectively controlling the strictness of the matching.
- **scorer**: Specifies the similarity scoring function used for matching (e.g., `fuzz.WRatio`).


# Example

```python
query = input("type Chinese word")
res_cn2dx = search_by_chinese(query)
print(res_cn2dx)
```

```python
query = input("type Dongxiang word")
res_dx2cn = search_by_dongxiang(query)
print(res_dx2cn)
```
